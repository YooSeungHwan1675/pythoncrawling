```python
ttalung = request.json()
ttalung["realtimeList"]
```

```python
ttalung_df = pd.DataFrame(ttalung["realtimeList"])
ttalung_df
```

```python
ttalung_df_map = ttalung_df[
    [
        'stationImgFileName', 'stationId', 'stationName', 'stationLongitude', 'stationLatitude', 'rackTotCnt', 'parkingBikeTotCnt', 'parkingQRBikeCnt',
        'parkingELECBikeCnt','stationSeCd','mode'
        
    ]
]
```

```python
ttalung_df_map['stationLongitude'] = ttalung_df_map['stationLongitude'].astype(float)
ttalung_df_map['stationLatitude'] = ttalung_df_map['stationLatitude'].astype(float)

ttalungsMap = folium.Map(location=[ttalung_df_map['stationLatitude'].mean(), ttalung_df_map['stationLongitude'].mean()], zoom_start=12)

for index, data in ttalung_df_map.iterrows():
    popup = folium.Popup('{} -- {}'.format(data['stationId'], data['stationName']), max_width=300)
    folium.Marker(location=[data['stationLatitude'], data['stationLongitude']], popup=popup).add_to(ttalungsMap)
ttalungsMap.save('./ttalungs_map.html')
ttalungsMap

```
